```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from imitation.data.types import Trajectory
from typing import List, Tuple
import logging

# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

class GameDataset(Dataset):
    def __init__(self, observations: np.ndarray, actions: np.ndarray):
        self.observations = torch.FloatTensor(observations)
        self.actions = torch.LongTensor(actions)
        
    def __len__(self):
        return len(self.actions)
    
    def __getitem__(self, idx):
        return self.observations[idx], self.actions[idx]

class BCPolicy(nn.Module):
    def __init__(self, observation_dim: int, action_dim: int, hidden_dim: int = 64):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(observation_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, action_dim)
        )
        
    def forward(self, x):
        return self.network(x)

def load_and_preprocess_data(data_file: str) -> Tuple[np.ndarray, np.ndarray]:
    """Load and preprocess the game data."""
    try:
        data = pd.read_csv(data_file)
        logger.info(f"Successfully loaded data from {data_file}")
        
        # Extract observations and actions
        observations = data.iloc[:, :8].values  # First 8 columns
        actions = data['player2_action'].values
        
        # Normalize observations
        observations = (observations - observations.mean(axis=0)) / (observations.std(axis=0) + 1e-8)
        
        return observations, actions
    
    except Exception as e:
        logger.error(f"Error loading data: {str(e)}")
        raise

def create_trajectories(observations: np.ndarray, actions: np.ndarray) -> List[Trajectory]:
    """Create trajectory objects for imitation learning."""
    if len(observations) != len(actions) + 1:
        observations = observations[:-1]  # Remove the last observation if lengths don't match
        logger.warning("Removed last observation to align lengths")
    
    terminals = [False] * (len(actions) - 1) + [True]
    return [Trajectory(obs=observations, acts=actions, infos=None, terminal=terminals)]

def train_bc_model(
    observations: np.ndarray,
    actions: np.ndarray,
    epochs: int = 100,
    batch_size: int = 64,
    learning_rate: float = 1e-3,
    validation_split: float = 0.2
):
    """Train the behavior cloning model."""
    # Create datasets
    dataset_size = len(actions)
    val_size = int(dataset_size * validation_split)
    train_size = dataset_size - val_size
    
    train_obs = observations[:train_size]
    train_acts = actions[:train_size]
    val_obs = observations[train_size:]
    val_acts = actions[train_size:]
    
    train_dataset = GameDataset(train_obs, train_acts)
    val_dataset = GameDataset(val_obs, val_acts)
    
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)
    
    # Initialize model and optimizer
    model = BCPolicy(observation_dim=8, action_dim=len(np.unique(actions)))
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    criterion = nn.CrossEntropyLoss()
    
    # Training loop
    best_val_loss = float('inf')
    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        for batch_obs, batch_acts in train_loader:
            optimizer.zero_grad()
            outputs = model(batch_obs)
            loss = criterion(outputs, batch_acts)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
        
        # Validation
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for batch_obs, batch_acts in val_loader:
                outputs = model(batch_obs)
                loss = criterion(outputs, batch_acts)
                val_loss += loss.item()
        
        # Logging
        if (epoch + 1) % 10 == 0:
            logger.info(f"Epoch [{epoch+1}/{epochs}], "
                       f"Train Loss: {train_loss/len(train_loader):.4f}, "
                       f"Val Loss: {val_loss/len(val_loader):.4f}")
        
        # Save best model
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), 'best_bc_model.pth')
    
    return model

def main():
    # Load and preprocess data
    observations, actions = load_and_preprocess_data("data.csv")
    logger.info(f"Loaded {len(observations)} observations and {len(actions)} actions")
    
    # Create trajectories for imitation learning
    demonstrations = create_trajectories(observations, actions)
    logger.info(f"Created {len(demonstrations)} demonstration trajectory")
    
    # Train the model
    model = train_bc_model(observations, actions)
    logger.info("Training completed successfully")
    
    return model, demonstrations

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from imitation.data.types import Trajectory
from typing import List, Tuple
import logging

# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

class GameDataset(Dataset):
    def __init__(self, observations: np.ndarray, actions: np.ndarray):
        self.observations = torch.FloatTensor(observations)
        self.actions = torch.LongTensor(actions)
        
    def __len__(self):
        return len(self.actions)
    
    def __getitem__(self, idx):
        return self.observations[idx], self.actions[idx]

class BCPolicy(nn.Module):
    def __init__(self, observation_dim: int, action_dim: int, hidden_dim: int = 64):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(observation_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, action_dim)
        )
        
    def forward(self, x):
        return self.network(x)

def load_and_preprocess_data(data_file: str) -> Tuple[np.ndarray, np.ndarray]:
    """Load and preprocess the game data."""
    try:
        data = pd.read_csv(data_file)
        logger.info(f"Successfully loaded data from {data_file}")
        
        # Extract observations and actions
        observations = data.iloc[:, :8].values  # First 8 columns
        actions = data['player2_action'].values
        
        # Normalize observations
        observations = (observations - observations.mean(axis=0)) / (observations.std(axis=0) + 1e-8)
        
        return observations, actions
    
    except Exception as e:
        logger.error(f"Error loading data: {str(e)}")
        raise

def create_trajectories(observations: np.ndarray, actions: np.ndarray) -> List[Trajectory]:
    """Create trajectory objects for imitation learning."""
    # Add an extra observation at the end by duplicating the last one
    extra_obs = np.vstack([observations, observations[-1:]])
    
    # Now we should have len(extra_obs) = len(actions) + 1
    assert len(extra_obs) == len(actions) + 1, f"Observation length {len(extra_obs)} != {len(actions)} + 1"
    
    terminals = [False] * (len(actions) - 1) + [True]
    return [Trajectory(obs=extra_obs, acts=actions, infos=None, terminal=terminals)]

def train_bc_model(
    observations: np.ndarray,
    actions: np.ndarray,
    epochs: int = 100,
    batch_size: int = 64,
    learning_rate: float = 1e-3,
    validation_split: float = 0.2
):
    """Train the behavior cloning model."""
    # Create datasets
    dataset_size = len(actions)
    val_size = int(dataset_size * validation_split)
    train_size = dataset_size - val_size
    
    train_obs = observations[:train_size]
    train_acts = actions[:train_size]
    val_obs = observations[train_size:]
    val_acts = actions[train_size:]
    
    train_dataset = GameDataset(train_obs, train_acts)
    val_dataset = GameDataset(val_obs, val_acts)
    
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)
    
    # Initialize model and optimizer
    model = BCPolicy(observation_dim=8, action_dim=len(np.unique(actions)))
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    criterion = nn.CrossEntropyLoss()
    
    # Training loop
    best_val_loss = float('inf')
    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        for batch_obs, batch_acts in train_loader:
            optimizer.zero_grad()
            outputs = model(batch_obs)
            loss = criterion(outputs, batch_acts)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
        
        # Validation
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for batch_obs, batch_acts in val_loader:
                outputs = model(batch_obs)
                loss = criterion(outputs, batch_acts)
                val_loss += loss.item()
        
        # Logging
        if (epoch + 1) % 10 == 0:
            logger.info(f"Epoch [{epoch+1}/{epochs}], "
                       f"Train Loss: {train_loss/len(train_loader):.4f}, "
                       f"Val Loss: {val_loss/len(val_loader):.4f}")
        
        # Save best model
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), 'best_bc_model.pth')
    
    return model

# Load and preprocess data
observations, actions = load_and_preprocess_data("data.csv")
logger.info(f"Loaded {len(observations)} observations and {len(actions)} actions")

# Create trajectories for imitation learning
demonstrations = create_trajectories(observations, actions)
logger.info(f"Created {len(demonstrations)} demonstration trajectory")

# Train the model
model = train_bc_model(observations, actions)
logger.info("Training completed successfully")
```
